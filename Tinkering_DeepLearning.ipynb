```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(x))
```

```python
sigmoid(0)
```

```python
x = np.linspace(-100, 100, 10000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, sigmoid(x))
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, x**3)
```
